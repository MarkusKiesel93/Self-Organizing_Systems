# Best combinations without constraints

```python
df = load_setup_1()
df = df[~df.use_constraint]
by_fit = {}
ITO = [10, 10, 30]
for fitness_function, ito in zip(FITNESS_FUNCTIONS, ITO):
    df_f = df[df.fitness_function == fitness_function].groupby(PARAM_COLUMNS[4:])[RESULT_COLUMNS].mean()
    df_f = df_f[(df_f.fitness == 1.0) & (df_f.optimum_reached == 1.0) & (df_f.iterations_to_opt < ito)]
    by_fit[fitness_function] = df_f
```

```python
for fitness_function in FITNESS_FUNCTIONS:
    print(fitness_function)
    df_sub = by_fit[fitness_function]
    print('Top performing params:')
    display(df_sub)
    print('Best param combination:')
    display(df_sub[df_sub.iterations_to_opt == df_sub.iterations_to_opt.min()])
```
